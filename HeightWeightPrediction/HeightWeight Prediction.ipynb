```python
data_women = data2[data2['Gender'] == 'Female']
data_women.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.hist(data_men['Height'], 
         histtype='stepfilled', 
         color='blue', 
         linewidth = 0.5, 
         edgecolor = 'black', 
         alpha = 0.5,
         density = True)

ax1.hist(data_women['Height'], 
         histtype='stepfilled', 
         color='red', 
         linewidth=0.5, 
         edgecolor = 'black',
         alpha = 0.5,
         density = True)

ax1.legend(['Men', 'Women']), ax1.set_title('Men and Women Heights')

ax2.hist(data_men['Weight'], 
         histtype='stepfilled', 
         color='blue', 
         linewidth = 0.5, 
         edgecolor = 'black', 
         alpha = 0.5, 
         density = True)

ax2.hist(data_women['Weight'], 
         histtype='stepfilled', 
         color='red', 
         linewidth=0.5, 
         edgecolor = 'black',
         alpha = 0.5, 
         density = True)

ax2.legend(['Men', 'Women']), ax2.set_title('Men and Women Weights')
```

```python
from sklearn.linear_model import LinearRegression

def plot_regression(dataset, gender):
    df = pd.DataFrame(data = dataset)
    df.plot.scatter(x = 'Weight', 
                        y = 'Height', 
                        edgecolor = 'black', 
                        linewidth = 0.2, 
                        title = gender + ' Height vs Weight Regression')

    X = df['Weight']
    Y = df['Height']
    X=X.values.reshape(len(X),1)
    Y=Y.values.reshape(len(Y),1)

    regr = LinearRegression()
    regr.fit(X,Y)
    plt.plot(X, regr.predict(X), color='red',linewidth=2)

plot_regression(data_men, 'Men')
plot_regression(data_women, 'Women')
```

```python
# weight = input('Input a Weight (lbs): ')
# weight = float(weight)
# prediction = regr.predict(weight)
# prediction = float(prediction)
# print("The predicted height is ", prediction, "inches")

def output_pred(dataset):
        df = pd.DataFrame(data = dataset)
        X = df['Weight']
        Y = df['Height']
        X=X.values.reshape(len(X),1)
        Y=Y.values.reshape(len(Y),1)
        regr = LinearRegression()
        regr.fit(X,Y)
        weight = input('Input a Weight (lbs): ')
        weight = float(weight)
        prediction = regr.predict(weight)
        prediction = float(prediction)
        print("The normal height for this weight is ", prediction, "inches")

def predict_height():
    gender = input("Gender (Male or Female): ")
#     gender = string(gender)
    if (gender.lower() == "male"):
        output_pred(data_men)
    elif(gender.lower() == "female"):
        output_pred(data_women)
    else:
        print("Not a valid gender")
```
